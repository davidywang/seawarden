```python
from imageai.Detection.Custom import CustomObjectDetection
import os
import pandas as pd
import shutil

json_path = "./detection_config.json"
#model_path = "/Users/Zack/0_imageai/resnet50_coco_best_v2.0.1.h5" #untrained model
#model_path = "/Users/Zack/0_imageai/detection_model-ex-094--loss-0028.326.h5" #imageai/google colab trained
model_path = "/Users/Zack/0_imageai/detection_model-ex-083--loss-0003.806.h5"
#model_path = "/Users/Zack/0_imageai/detection_model-ex-083--loss-0003.806.h5"

#input_dir = "/Users/Zack/0_bing/images_1km_2020-02-08/"
input_dir = "/Users/Zack/0_bing/images_1km_2019-10-13/"
output_dir = "./t/"

select_dir1 = "./d/" #copy original images
select_dir2 = "./d_a/" #copy annotated images 
```

```python
detector = CustomObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath(model_path) 
detector.setJsonPath(json_path)
detector.loadModel()
```

```python
img_list, detect_list = [], []

for img in os.listdir(input_dir):
    input_path= os.path.join(input_dir, img)
    output_path = os.path.join(output_dir, img)
    detections = detector.detectObjectsFromImage(input_image=input_path, output_image_path=output_path)
    detect_list.append(len(detections))
    img_list.append(img)
    print("img: %s\t, num_detections: %d" % (img, len(detections)))
```

```python
#select images with detections 
df = pd.DataFrame({'image': img_list,'detections': detect_list})
select = df['detections'] > 1
#select = df['detections'] > 4
select = df[select]
select

#select.to_csv('detections.csv', index = None, header=True)
```

```python
#copy images with detections
select_list = select['image'].tolist()

for img in select_list:
    org_img = os.path.join(input_dir, img)
    annotated = os.path.join(output_dir, img)
    shutil.copy(org_img, select_dir1)
    shutil.copy(annotated, select_dir2)
```
