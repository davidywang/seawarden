```python
from imageai.Detection.Custom import CustomObjectDetection
import os
import pandas as pd
import shutil

json_path = "./detection_config.json"
model_path = "/Users/Zack/0_thesis_imageai/detection_model-ex-094--loss-0028.326.h5"

input_dir = "/Users/Zack/0_thesis_bing/images_1km_2020-02-08/"
output_dir = "./test"

select_dir1 = "./detections/" #copy original images
select_dir2 = "./detections_annotated" #copy annotated images 
```

```python
detector = CustomObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath(model_path) 
detector.setJsonPath(json_path)
detector.loadModel()
```

```python
img_list, detect_list = [], []

for img in os.listdir(input_dir):
    input_path= os.path.join(input_dir, img)
    output_path = os.path.join(output_dir, img)
    detections = detector.detectObjectsFromImage(input_image=input_path, output_image_path=output_path)
    detect_list.append(len(detections))
    img_list.append(img)
    print("img: %s\t, num_detections: %d" % (img, len(detections)))
```

```python
#select images with detections 
df = pd.DataFrame({'image': img_list,'detections': detect_list})
select = df['detections'] > 1
#select = df['detections'] > 4
select = df[select]
select

#select.to_csv('detections.csv', index = None, header=True)
```
