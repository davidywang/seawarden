```python
import geopandas as gpd

grid = gpd.read_file('/Users/Zack/0_GIS_Greece/AOI/greece_aoi_05deg.shp')
mask = gpd.read_file('/Users/Zack/0_GIS_Greece/dem_eu/eu_dem_v11_mask.shp')
mask = mask.to_crs({'init': 'epsg:2100'})

print(grid.crs)
print(mask.crs)

ax = grid.plot(linewidth=.5, edgecolor = 'black', facecolor = 'none')
mask.plot(linewidth=.5, edgecolor = 'none', facecolor = 'grey', ax=ax)
```

```python
%%time 
matches = grid.intersects(mask.unary_union)
grid_matches = grid.loc[matches]
print(len(grid))
print(len(grid_matches))

grid_matches.plot(linewidth=.5, edgecolor = 'black', facecolor = 'none')
```
