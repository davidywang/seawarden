```python
#load csv containing coordinates of polygon corners
import pandas as pd
file = '/Users/Zack/0_thesis_bing/aquaculture_targets.csv'
targets = pd.read_csv(file, delimiter=',', header=0)
print(len(targets))
targets.head()
```

```python
topleft = targets.drop(targets.index[0])
topleft = topleft.iloc[::5]
print(len(topleft))
topleft.head()
```

```python
bright = targets.drop(targets.index[0:3])
bright = bright.iloc[::5]
print(len(bright))
bright.head()
```

```python
#single location extractor
from aerialImageRetrieval_2 import AerialImageRetrieval
x=239
lat1=topleft['y'].tolist()[x]
lon1=topleft['x'].tolist()[x]
lat2=bright['y'].tolist()[x]
lon2=bright['x'].tolist()[x]

print(topleft['y'].tolist()[x])
print(topleft['x'].tolist()[x])
print(bright['y'].tolist()[x])
print(bright['x'].tolist()[x])

imgretrieval = AerialImageRetrieval(lat1, lon1, lat2, lon2)
if imgretrieval.max_resolution_imagery_retrieval():
    print("Successfully retrieved the image with maximum resolution!")
else:
    print("Cannot retrieve the desired image! (Possible reason: expected tile image does not exist.)")
```

```python
# create lists
lat1_list=topleft['y'].tolist()
lon1_list=topleft['x'].tolist()
lat2_list=bright['y'].tolist()
lon2_list=bright['x'].tolist()
from aerialImageRetrieval_2 import AerialImageRetrieval

for i in range(len(lat1_list)):
    lat1=round(lat1_list[i],4)
    lon1=round(lon1_list[i],4)
    lat2=round(lat2_list[i],4)
    lon2=round(lon2_list[i],4)
    print(lat1, lon1, lat2, lon2)
    
    imgretrieval = AerialImageRetrieval(lat1, lon1, lat2, lon2)
    if imgretrieval.max_resolution_imagery_retrieval():
        print(lat1, lon1, lat2, lon2, "Successfully retrieved the image with maximum resolution!")
    else:
        print(lat1, lon1, lat2, lon2, "Cannot retrieve the desired image! (Possible reason: expected tile image does not exist.)")
```

```python
#add farm ID to filename (quick fix)
import os 

folder = "/Users/Zack/0_thesis_bing/images_2019-10-13/"
files = os.listdir(folder)
#print(files)

#sort files by date: newest to oldest
search_dir = folder
os.chdir(search_dir)
files = filter(os.path.isfile, os.listdir(search_dir))
files = [os.path.join(f) for f in files] 
files.sort(key=lambda x: os.path.getmtime(x))
print(len(files))

#rename with farm ID
for index, i in enumerate(files):
    org = (i[11:])
    name = (str(index))
    dst = (name + org)
    src = (i)
    os.rename(src, dst)
```
