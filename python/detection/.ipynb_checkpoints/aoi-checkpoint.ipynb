```python
import pandas as pd
import geopandas as gpd

#load spatial data
greece = gpd.read_file('/Users/Zack/0_GIS_Greece/GRC_adm/GRC_adm0.shp')
eez = gpd.read_file('/Users/Zack/0_GIS_Greece/EEZ/greece_EEZ.shp')

eez_prep = gpd.read_file('/Users/Zack/0_GIS_Greece/EEZ/eez_prep.shp') 
#created with geojson.io

mask = gpd.read_file('/Users/Zack/0_GIS_Greece/dem_eu/eu_dem_v11_mask.shp') 
#will obtain mask directly when #gdal is resolved

print('greece:', greece.crs)
print('eez:', eez.crs)
print('eez_prep:', eez_prep.crs)
print('mask:', mask.crs)
```

```python
#create envelope

#generate envelopes
eez_e = gpd.GeoDataFrame(geometry = eez.envelope)
greece_e = gpd.GeoDataFrame(geometry = greece.envelope)

#combine envelopes
merge = gpd.overlay(greece_e, eez_e, how = 'union')
merge['Dissolve'] = 0

#dissolve envelopes and make new envelope
envelope = merge.dissolve('Dissolve')
envelope = gpd.GeoDataFrame(geometry = envelope.envelope)
#print('envelope', envelope.crs)

#change projection
#envelope = envelope.to_crs({'init': 'epsg:4326'})
#print('envelope', envelope.crs)

#plot
ax = eez_e.plot(linewidth=.5, edgecolor = 'red', facecolor = 'none')
greece_e.plot(linewidth=.5, edgecolor = 'blue', facecolor = 'none', ax=ax)
envelope.plot(linewidth=1, edgecolor = 'green', facecolor = 'none', ax=ax)

#export
#envelope.to_file("/Users/Zack/0_GIS_Greece/aoi/greece_envelope.shp")
#envelope.to_file("/Users/Zack/Desktop/greece_envelope.geojson", driver="GeoJSON")
```

```python
merge = gpd.overlay(eez, eez_prep, how = 'union')
merge.plot()
```

```python
%%time 
#convert eez to an outline
merge = gpd.overlay(eez, eez_prep, how = 'union')
merge['Dissolve'] = 0
eez_dis = merge.dissolve('Dissolve')
eez_dis.plot(linewidth=.5, edgecolor = 'blue', facecolor = 'none')
```

```python
%%time 
#combine eez and greece to create aoi
merge = gpd.overlay(eez_dis, greece, how = 'union')
merge['Dissolve'] = 0
aoi = merge.dissolve('Dissolve')
aoi.plot(linewidth=.5, edgecolor = 'blue', facecolor = 'none')
```

```python
#export aoi
aoi.to_file('/Users/Zack/0_GIS_Greece/aoi/aoi.shp')
```

```python
%%time 
#generate 50m buffer of land mask
mask = mask.to_crs({'init': 'epsg:2100'})
mask_50m = mask.buffer(50)
mask_50m = gpd.GeoDataFrame(geometry = mask_50m)
mask_50m.plot()
```

```python
%%time
#generate 5km buffer of land mask
mask_5km = mask.buffer(5000)
mask_5km = gpd.GeoDataFrame(geometry = mask_5km)
mask_5km.plot()
```

```python
%%time
#union 50m land mask to aoi to create water_50m
mask_50m = mask_50m.to_crs({'init': 'epsg:4326'})
aoi = mask_50m.to_crs({'init': 'epsg:4326'})

merge = gpd.overlay(aoi, mask_50m, how = 'intersection')
merge.plot(linewidth=.5, edgecolor = 'blue', facecolor = 'none')
```
