```python
#https://windrose.readthedocs.io/en/latest/usage.html
#https://stackoverflow.com/questions/50821406/custom-scaling-of-wind-rose-python
#https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html

import os 
import pandas as pd
from windrose import WindroseAxes
from matplotlib import pyplot as plt
import matplotlib.cm as cm
import numpy as np
```

```python
#load CSVs
folder = '../data_current/current_csv/'
files = os.listdir(folder)
print(len(files))
files
```

```python
#plot function
def plot_rose(wd, ws, name):
    ax = WindroseAxes.from_ax()
    ax.bar(wd, ws, normed=True, opening=.75, edgecolor='white', cmap=cm.summer) #_r)
    ax.set_yticks(np.arange(5, 25, step=5))
    ax.set_yticklabels(np.arange(5, 25, step=5))
    ax.set_legend(loc='center', bbox_to_anchor=(.5, -.1), ncol = 6, prop={'size': 12})
    #ax.set_legend(loc=3,ncol=6, mode="expand", borderaxespad=0.)
    plt.title('Farm Site ' + name + ': Ocean Current and Direction')
```

```python
#plot tester
i=5
file = folder + files[i]
csv = pd.read_csv(file)
ws = csv['speed']
wd = csv['dir']
name = files[i].split("_")[0]
plot_rose(wd, ws, name)
```

```python
out_dir = '../data_current/current_plots/'

for i in range(len(files)):
    file = folder + files[i]
    csv = pd.read_csv(file)
    plot_rose(wd=csv['dir'], ws=csv['speed'], name = files[i].split("_")[0]) 
    filename = out_dir + files[i].split(".")[0] + '_plot.png'
    plot = plt.savefig(filename)
    plt.close(plot)
```
